```python
# Complete APTOS Federated Learning with InceptionV3
# Generates: Accuracy, F1, AUC, Time, Expert Entropy, Parameters, Communication Costs, ROC, Confusion Matrix

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Subset, WeightedRandomSampler
from torchvision import transforms, models
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (accuracy_score, f1_score, recall_score,
                            confusion_matrix, classification_report,
                            roc_curve, auc)
from sklearn.preprocessing import label_binarize # Import label_binarize from preprocessing
import os
import time
import copy
import gc

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
```

```python
# ==== Inception-ResNet-V2 Architecture Components ====

class ConvBNReLU(nn.Module):
    """Convolution + BatchNorm + ReLU"""
    def __init__(self, in_channels, out_channels, kernel_size, stride=1, padding=0):
        super(ConvBNReLU, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size, stride, padding, bias=False)
        self.bn = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        return self.relu(self.bn(self.conv(x)))


class InceptionResNetA(nn.Module):
    """Inception-ResNet-A block"""
    def __init__(self, in_channels, scale=0.17):
        super(InceptionResNetA, self).__init__()
        self.scale = scale

        self.branch1x1 = ConvBNReLU(in_channels, 32, 1)

        self.branch3x3_1 = ConvBNReLU(in_channels, 32, 1)
        self.branch3x3_2 = ConvBNReLU(32, 32, 3, padding=1)

        self.branch3x3dbl_1 = ConvBNReLU(in_channels, 32, 1)
        self.branch3x3dbl_2 = ConvBNReLU(32, 48, 3, padding=1)
        self.branch3x3dbl_3 = ConvBNReLU(48, 64, 3, padding=1)

        self.conv = nn.Conv2d(128, in_channels, 1)
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        branch1x1 = self.branch1x1(x)

        branch3x3 = self.branch3x3_1(x)
        branch3x3 = self.branch3x3_2(branch3x3)

        branch3x3dbl = self.branch3x3dbl_1(x)
        branch3x3dbl = self.branch3x3dbl_2(branch3x3dbl)
        branch3x3dbl = self.branch3x3dbl_3(branch3x3dbl)

        out = torch.cat([branch1x1, branch3x3, branch3x3dbl], dim=1)
        out = self.conv(out)
        out = out * self.scale + x
        out = self.relu(out)
        return out


class InceptionResNetB(nn.Module):
    """Inception-ResNet-B block"""
    def __init__(self, in_channels, scale=0.10):
        super(InceptionResNetB, self).__init__()
        self.scale = scale

        self.branch1x1 = ConvBNReLU(in_channels, 192, 1)

        self.branch7x7_1 = ConvBNReLU(in_channels, 128, 1)
        self.branch7x7_2 = ConvBNReLU(128, 160, (1, 7), padding=(0, 3))
        self.branch7x7_3 = ConvBNReLU(160, 192, (7, 1), padding=(3, 0))

        self.conv = nn.Conv2d(384, in_channels, 1)
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        branch1x1 = self.branch1x1(x)

        branch7x7 = self.branch7x7_1(x)
        branch7x7 = self.branch7x7_2(branch7x7)
        branch7x7 = self.branch7x7_3(branch7x7)

        out = torch.cat([branch1x1, branch7x7], dim=1)
        out = self.conv(out)
        out = out * self.scale + x
        out = self.relu(out)
        return out


class InceptionResNetC(nn.Module):
    """Inception-ResNet-C block"""
    def __init__(self, in_channels, scale=0.20):
        super(InceptionResNetC, self).__init__()
        self.scale = scale

        self.branch1x1 = ConvBNReLU(in_channels, 192, 1)

        self.branch3x3_1 = ConvBNReLU(in_channels, 192, 1)
        self.branch3x3_2 = ConvBNReLU(192, 224, (1, 3), padding=(0, 1))
        self.branch3x3_3 = ConvBNReLU(224, 256, (3, 1), padding=(1, 0))

        self.conv = nn.Conv2d(448, in_channels, 1)
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        branch1x1 = self.branch1x1(x)

        branch3x3 = self.branch3x3_1(x)
        branch3x3 = self.branch3x3_2(branch3x3)
        branch3x3 = self.branch3x3_3(branch3x3)

        out = torch.cat([branch1x1, branch3x3], dim=1)
        out = self.conv(out)
        out = out * self.scale + x
        out = self.relu(out)
        return out


class ReductionA(nn.Module):
    """Reduction-A (35x35 to 17x17)"""
    def __init__(self, in_channels, k=256, l=256, m=384, n=384):
        super(ReductionA, self).__init__()
        self.branch3x3 = ConvBNReLU(in_channels, n, 3, stride=2)

        self.branch3x3dbl_1 = ConvBNReLU(in_channels, k, 1)
        self.branch3x3dbl_2 = ConvBNReLU(k, l, 3, padding=1)
        self.branch3x3dbl_3 = ConvBNReLU(l, m, 3, stride=2)

        self.branch_pool = nn.MaxPool2d(3, stride=2)

    def forward(self, x):
        branch3x3 = self.branch3x3(x)

        branch3x3dbl = self.branch3x3dbl_1(x)
        branch3x3dbl = self.branch3x3dbl_2(branch3x3dbl)
        branch3x3dbl = self.branch3x3dbl_3(branch3x3dbl)

        branch_pool = self.branch_pool(x)

        return torch.cat([branch3x3, branch3x3dbl, branch_pool], dim=1)


class ReductionB(nn.Module):
    """Reduction-B (17x17 to 8x8)"""
    def __init__(self, in_channels):
        super(ReductionB, self).__init__()
        self.branch3x3_1 = ConvBNReLU(in_channels, 256, 1)
        self.branch3x3_2 = ConvBNReLU(256, 384, 3, stride=2)

        self.branch3x3dbl_1 = ConvBNReLU(in_channels, 256, 1)
        self.branch3x3dbl_2 = ConvBNReLU(256, 288, 3, stride=2)

        self.branch3x3dbl2_1 = ConvBNReLU(in_channels, 256, 1)
        self.branch3x3dbl2_2 = ConvBNReLU(256, 288, 3, padding=1)
        self.branch3x3dbl2_3 = ConvBNReLU(288, 320, 3, stride=2)

        self.branch_pool = nn.MaxPool2d(3, stride=2)

    def forward(self, x):
        branch3x3 = self.branch3x3_1(x)
        branch3x3 = self.branch3x3_2(branch3x3)

        branch3x3dbl = self.branch3x3dbl_1(x)
        branch3x3dbl = self.branch3x3dbl_2(branch3x3dbl)

        branch3x3dbl2 = self.branch3x3dbl2_1(x)
        branch3x3dbl2 = self.branch3x3dbl2_2(branch3x3dbl2)
        branch3x3dbl2 = self.branch3x3dbl2_3(branch3x3dbl2)

        branch_pool = self.branch_pool(x)

        return torch.cat([branch3x3, branch3x3dbl, branch3x3dbl2, branch_pool], dim=1)


class InceptionResNetV2(nn.Module):
    """Inception-ResNet-V2 for APTOS classification"""
    def __init__(self, num_classes=5, dropout_rate=0.4):
        super(InceptionResNetV2, self).__init__()

        # Stem
        self.conv2d_1a = ConvBNReLU(3, 32, 3, stride=2)
        self.conv2d_2a = ConvBNReLU(32, 32, 3)
        self.conv2d_2b = ConvBNReLU(32, 64, 3, padding=1)
        self.maxpool_3a = nn.MaxPool2d(3, stride=2)
        self.conv2d_3b = ConvBNReLU(64, 80, 1)
        self.conv2d_4a = ConvBNReLU(80, 192, 3)
        self.maxpool_5a = nn.MaxPool2d(3, stride=2)

        # Inception-ResNet-A (5x)
        self.mixed_5b = ConvBNReLU(192, 320, 1)
        self.repeat_a = nn.Sequential(*[InceptionResNetA(320) for _ in range(5)])

        # Reduction-A
        self.mixed_6a = ReductionA(320)

        # Inception-ResNet-B (10x)
        self.repeat_b = nn.Sequential(*[InceptionResNetB(1088) for _ in range(10)])

        # Reduction-B
        self.mixed_7a = ReductionB(1088)

        # Inception-ResNet-C (5x)
        self.repeat_c = nn.Sequential(*[InceptionResNetC(2080) for _ in range(5)])

        # Final pooling and classifier
        self.avgpool = nn.AdaptiveAvgPool2d((1, 1))
        self.dropout = nn.Dropout(dropout_rate)

        self.classifier = nn.Sequential(
            nn.Dropout(dropout_rate),
            nn.Linear(2080, 1024),
            nn.ReLU(inplace=True),
            nn.BatchNorm1d(1024),
            nn.Dropout(dropout_rate),
            nn.Linear(1024, 512),
            nn.ReLU(inplace=True),
            nn.BatchNorm1d(512),
            nn.Dropout(dropout_rate / 2),
            nn.Linear(512, num_classes)
        )

        self.is_moe = False
        self._initialize_weights()

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)

    def forward(self, x):
        # Stem
        x = self.conv2d_1a(x)
        x = self.conv2d_2a(x)
        x = self.conv2d_2b(x)
        x = self.maxpool_3a(x)
        x = self.conv2d_3b(x)
        x = self.conv2d_4a(x)
        x = self.maxpool_5a(x)

        # Inception-ResNet-A
        x = self.mixed_5b(x)
        x = self.repeat_a(x)

        # Reduction-A
        x = self.mixed_6a(x)

        # Inception-ResNet-B
        x = self.repeat_b(x)

        # Reduction-B
        x = self.mixed_7a(x)

        # Inception-ResNet-C
        x = self.repeat_c(x)

        # Global pooling and classifier
        x = self.avgpool(x)
        x = x.view(x.size(0), -1)
        x = self.classifier(x)

        return x


# ==== MOE Components ====
class ExpertNetwork(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, dropout_rate=0.3):
        super(ExpertNetwork, self).__init__()
        self.network = nn.Sequential(
            nn.LayerNorm(input_dim),
            nn.Linear(input_dim, hidden_dim),
            nn.GELU(),
            nn.Dropout(dropout_rate),
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.GELU(),
            nn.Dropout(dropout_rate),
            nn.Linear(hidden_dim // 2, output_dim)
        )

    def forward(self, x):
        return self.network(x)


class GatingNetwork(nn.Module):
    def __init__(self, input_dim, num_experts, dropout_rate=0.1):
        super(GatingNetwork, self).__init__()
        self.gate = nn.Sequential(
            nn.LayerNorm(input_dim),
            nn.Linear(input_dim, input_dim // 2),
            nn.GELU(),
            nn.Dropout(dropout_rate),
            nn.Linear(input_dim // 2, num_experts),
            nn.Softmax(dim=-1)
        )

    def forward(self, x):
        return self.gate(x)


class MixtureOfExperts(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, num_experts=4, dropout_rate=0.3, top_k=2):
        super(MixtureOfExperts, self).__init__()
        self.num_experts = num_experts
        self.top_k = top_k
        self.experts = nn.ModuleList([ExpertNetwork(input_dim, hidden_dim, output_dim, dropout_rate) for _ in range(num_experts)])
        self.gate = GatingNetwork(input_dim, num_experts, dropout_rate)

    def forward(self, x):
        gate_weights = self.gate(x)
        top_k_weights, top_k_indices = torch.topk(gate_weights, self.top_k, dim=-1)
        top_k_weights = top_k_weights / top_k_weights.sum(dim=-1, keepdim=True)

        expert_outputs = torch.stack([expert(x) for expert in self.experts], dim=1)
        batch_size, output_dim = x.size(0), expert_outputs.size(-1)
        final_output = torch.zeros(batch_size, output_dim, device=x.device)

        for i in range(self.top_k):
            expert_idx = top_k_indices[:, i]
            expert_weight = top_k_weights[:, i].unsqueeze(-1)
            selected_outputs = expert_outputs[torch.arange(batch_size), expert_idx]
            final_output += expert_weight * selected_outputs

        return final_output, gate_weights


class InceptionResNetV2MOE(nn.Module):
    """Inception-ResNet-V2 with MOE classifier"""
    def __init__(self, num_classes=5, dropout_rate=0.4, num_experts=4):
        super(InceptionResNetV2MOE, self).__init__()

        # Build base model
        base_model = InceptionResNetV2(num_classes=num_classes, dropout_rate=0.0)

        # Copy all layers except classifier
        self.conv2d_1a = base_model.conv2d_1a
        self.conv2d_2a = base_model.conv2d_2a
        self.conv2d_2b = base_model.conv2d_2b
        self.maxpool_3a = base_model.maxpool_3a
        self.conv2d_3b = base_model.conv2d_3b
        self.conv2d_4a = base_model.conv2d_4a
        self.maxpool_5a = base_model.maxpool_5a
        self.mixed_5b = base_model.mixed_5b
        self.repeat_a = base_model.repeat_a
        self.mixed_6a = base_model.mixed_6a
        self.repeat_b = base_model.repeat_b
        self.mixed_7a = base_model.mixed_7a
        self.repeat_c = base_model.repeat_c
        self.avgpool = base_model.avgpool

        # MOE classifier
        self.classifier = nn.Sequential(
            nn.Dropout(dropout_rate),
            MixtureOfExperts(
                input_dim=2080,
                hidden_dim=1024,
                output_dim=num_classes,
                num_experts=num_experts,
                dropout_rate=dropout_rate,
                top_k=2
            )
        )

        self.is_moe = True

    def forward(self, x):
        # Same forward as base model up to pooling
        x = self.conv2d_1a(x)
        x = self.conv2d_2a(x)
        x = self.conv2d_2b(x)
        x = self.maxpool_3a(x)
        x = self.conv2d_3b(x)
        x = self.conv2d_4a(x)
        x = self.maxpool_5a(x)
        x = self.mixed_5b(x)
        x = self.repeat_a(x)
        x = self.mixed_6a(x)
        x = self.repeat_b(x)
        x = self.mixed_7a(x)
        x = self.repeat_c(x)
        x = self.avgpool(x)
        x = x.view(x.size(0), -1)

        # MOE classifier returns (logits, gate_weights)
        return self.classifier[1](x)


# ==== Loss Functions ====
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        ce_loss = nn.functional.cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = (1 - pt) ** self.gamma * ce_loss

        if self.alpha is not None:
            alpha_t = self.alpha[targets]
            focal_loss = alpha_t * focal_loss

        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        return focal_loss


class MOELoss(nn.Module):
    def __init__(self, base_criterion, load_balance_weight=0.01):
        super(MOELoss, self).__init__()
        self.base_criterion = base_criterion
        self.load_balance_weight = load_balance_weight

    def forward(self, outputs, targets):
        if isinstance(outputs, tuple):
            logits, gate_weights = outputs
            base_loss = self.base_criterion(logits, targets)
        else:
            logits = outputs
            base_loss = self.base_criterion(logits, targets)
            gate_weights = None

        total_loss = base_loss

        if gate_weights is not None and self.load_balance_weight > 0:
            num_experts = gate_weights.size(-1)
            mean_gate_weights = gate_weights.mean(dim=0)
            uniform_target = torch.ones_like(mean_gate_weights) / num_experts
            load_balance_loss = nn.functional.kl_div(torch.log(mean_gate_weights + 1e-8), uniform_target, reduction='batchmean')
            total_loss += self.load_balance_weight * load_balance_loss

        return total_loss


# ==== Dataset Setup ====
def setup_aptos_dataset_with_split(base_path, test_ratio=0.2, random_seed=42):
    np.random.seed(random_seed)
    torch.manual_seed(random_seed)

    # Inception-ResNet-V2 uses 299x299 input
    train_tfm = transforms.Compose([
        transforms.Resize((320, 320)),
        transforms.RandomResizedCrop(299, scale=(0.8, 1.0)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomVerticalFlip(p=0.5),
        transforms.RandomRotation(20),
        transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.1),
        transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    test_tfm = transforms.Compose([
        transforms.Resize((299, 299)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    class_folders = sorted([f for f in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, f))])
    print(f"Found classes: {class_folders}")

    all_samples = []
    class_to_idx = {class_name: idx for idx, class_name in enumerate(class_folders)}

    for class_name in class_folders:
        class_path = os.path.join(base_path, class_name)
        class_idx = class_to_idx[class_name]
        image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        print(f"Class '{class_name}': {len(image_files)} samples")
        for img_file in image_files:
            all_samples.append((os.path.join(class_path, img_file), class_idx))

    train_samples, test_samples = [], []
    class_counts = torch.zeros(len(class_folders))

    for class_idx, class_name in enumerate(class_folders):
        class_samples = [(path, label) for path, label in all_samples if label == class_idx]
        np.random.shuffle(class_samples)
        n_test = max(1, int(len(class_samples) * test_ratio))
        n_train = len(class_samples) - n_test
        train_samples.extend(class_samples[:n_train])
        test_samples.extend(class_samples[n_train:])
        class_counts[class_idx] = n_train

    np.random.shuffle(train_samples)
    np.random.shuffle(test_samples)

    class APTOSDataset(torch.utils.data.Dataset):
        def __init__(self, samples, transform=None):
            self.samples = samples
            self.transform = transform

        def __len__(self):
            return len(self.samples)

        def __getitem__(self, idx):
            from PIL import Image
            path, label = self.samples[idx]
            image = Image.open(path).convert('RGB')
            if self.transform:
                image = self.transform(image)
            return image, label

    train_ds = APTOSDataset(train_samples, transform=train_tfm)
    test_ds = APTOSDataset(test_samples, transform=test_tfm)
    train_ds.classes = test_ds.classes = class_folders

    class_weights = 1.0 / (class_counts + 1e-8)
    class_weights = class_weights / class_weights.sum() * len(class_folders)

    print(f"\nDataset prepared:")
    print(f"  Training samples: {len(train_ds)}")
    print(f"  Test samples: {len(test_ds)}")

    return train_ds, test_ds, class_folders, class_weights, len(class_folders)


def split_noniid_aptos(labels, n_clients=4, alpha=0.5):
    idxs = np.arange(len(labels))
    client_idx = [[] for _ in range(n_clients)]
    unique_classes = np.unique(labels)

    print(f"\nCreating Non-IID split with alpha={alpha}")
    for c in unique_classes:
        idx_c = idxs[labels == c]
        np.random.shuffle(idx_c)
        props = np.random.dirichlet([alpha] * n_clients)
        props = props / props.sum()
        splits = (props.cumsum()[:-1] * len(idx_c)).astype(int)
        splits = np.split(idx_c, splits)
        for i in range(n_clients):
            client_idx[i] += splits[i].tolist()

    for i in range(n_clients):
        np.random.shuffle(client_idx[i])

    return client_idx


# ==== Model Creation ====
def get_inceptionresnetv2_model(num_classes=5, dropout_rate=0.4, use_moe=False, num_experts=4):
    """Create Inception-ResNet-V2 model"""

    if use_moe:
        model = InceptionResNetV2MOE(
            num_classes=num_classes,
            dropout_rate=dropout_rate,
            num_experts=num_experts
        )
        print(f"Created Inception-ResNet-V2 with MOE ({num_experts} experts)")
    else:
        model = InceptionResNetV2(
            num_classes=num_classes,
            dropout_rate=dropout_rate
        )
        print("Created Inception-ResNet-V2 model")

    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

    print(f"Total parameters: {total_params:,}")
    print(f"Trainable parameters: {trainable_params:,}")
    print(f"Parameters (M): {total_params / 1e6:.2f}")
    print(f"Model size: {total_params * 4 / (1024**2):.2f} MB")

    return model


# ==== Communication Cost Tracker ====
class CommCostTracker:
    def __init__(self, model):
        self.param_count = sum(p.numel() for p in model.parameters())
        self.param_size_mb = self.param_count * 4 / (1024 ** 2)

    def get_detailed_costs(self):
        return {
            'parameters_m': self.param_count / 1e6,
            'uplink_mb': self.param_size_mb,
            'downlink_mb': self.param_size_mb,
            'total_comm_mb': self.param_size_mb * 2
        }


# ==== Training Function ====
def train_local_inceptionresnetv2(model, loader, epochs=5, lr=1e-4, weight_decay=1e-4, class_weights=None):
    """Training optimized for Inception-ResNet-V2"""
    model = model.to(device)
    model.train()
    start_time = time.time()

    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)

    base_criterion = FocalLoss(alpha=class_weights.to(device) if class_weights is not None else None, gamma=2.0)
    criterion = MOELoss(base_criterion, 0.01 if hasattr(model, 'is_moe') and model.is_moe else 0.0)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=lr*0.01)

    total_loss = 0
    total_samples = 0

    for epoch in range(epochs):
        for X, y in loader:
            X, y = X.to(device), y.to(device)
            optimizer.zero_grad()
            outputs = model(X)
            loss = criterion(outputs, y)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            total_loss += loss.item() * X.size(0)
            total_samples += X.size(0)
        scheduler.step()

    return copy.deepcopy(model.state_dict()), total_loss / total_samples, time.time() - start_time


# ==== Evaluation ====
def eval_model_enhanced(model, loader, class_names):
    model.eval()
    y_true, y_pred, y_prob = [], [], []
    total_loss = 0
    gate_weights_all = []
    criterion = nn.CrossEntropyLoss()

    with torch.no_grad():
        for X, y in loader:
            X, y = X.to(device), y.to(device)
            outputs = model(X)

            if hasattr(model, 'is_moe') and model.is_moe and isinstance(outputs, tuple):
                logits, gate_weights = outputs
                gate_weights_all.append(gate_weights.cpu())
            else:
                logits = outputs

            loss = criterion(logits, y)
            total_loss += loss.item() * X.size(0)
            probs = torch.softmax(logits, dim=1)
            y_true.extend(y.cpu().numpy())
            y_pred.extend(logits.argmax(dim=1).cpu().numpy())
            y_prob.extend(probs.cpu().numpy())

    y_true, y_pred, y_prob = np.array(y_true), np.array(y_pred), np.array(y_prob)
    accuracy = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average='weighted')

    try:
        y_true_bin = label_binarize(y_true, classes=range(len(class_names)))
        auc_scores = []
        for i in range(len(class_names)):
            if len(np.unique(y_true_bin[:, i])) > 1:
                fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_prob[:, i])
                auc_scores.append(auc(fpr, tpr))
            else:
                auc_scores.append(0.0)
        auc_roc = np.mean(auc_scores)
    except:
        auc_roc = 0

    results = {
        'accuracy': accuracy,
        'f1': f1,
        'auc_roc': auc_roc,
        'loss': total_loss / len(loader.dataset),
        'y_true': y_true,
        'y_pred': y_pred,
        'y_prob': y_prob
    }

    if gate_weights_all:
        gate_weights_all = torch.cat(gate_weights_all, dim=0)
        expert_usage = gate_weights_all.mean(dim=0).numpy()
        expert_entropy = -np.sum(expert_usage * np.log(expert_usage + 1e-8))
        results['expert_usage'] = expert_usage
        results['expert_entropy'] = expert_entropy

    return results


# ==== Aggregation Methods ====
def fedavg_enhanced(states, weights=None):
    if weights is None:
        weights = [1.0 / len(states)] * len(states)
    avg_state = {}
    for key in states[0].keys():
        if states[0][key].dtype in [torch.float32, torch.float64, torch.float16]:
            avg_state[key] = sum(weights[i] * states[i][key].float() for i in range(len(states)))
        else:
            avg_state[key] = states[0][key].clone()
    return avg_state


def fedprox_aggregation(states, global_state, mu=0.01):
    avg_state = fedavg_enhanced(states)
    for key in avg_state.keys():
        if key in global_state and avg_state[key].dtype in [torch.float32, torch.float64, torch.float16]:
            avg_state[key] = (1 - mu) * avg_state[key] + mu * global_state[key].float()
    return avg_state


def moe_aggregation(local_states, client_weights=None, num_experts=4):
    if client_weights is None:
        client_weights = [1.0 / len(local_states)] * len(local_states)

    shared_state = fedavg_enhanced([
        {k: v for k, v in state.items() if 'classifier' not in k}
        for state in local_states
    ], client_weights)

    expert_states = {}
    for expert_idx in range(num_experts):
        prefix = f'classifier.1.experts.{expert_idx}'
        expert_params = [{k.replace(prefix + '.', ''): v for k, v in state.items() if k.startswith(prefix)} for state in local_states]
        expert_params = [ep for ep in expert_params if ep]
        if expert_params:
            agg_expert = fedavg_enhanced(expert_params, client_weights)
            for k, v in agg_expert.items():
                expert_states[f'{prefix}.{k}'] = v

    gate_params = [{k.replace('classifier.1.gate.', ''): v for k, v in state.items() if k.startswith('classifier.1.gate.')} for state in local_states]
    gate_states = {}
    if gate_params and gate_params[0]:
        agg_gate = fedavg_enhanced(gate_params, client_weights)
        gate_states = {f'classifier.1.gate.{k}': v for k, v in agg_gate.items()}

    other_states = {}
    for state in local_states:
        for k, v in state.items():
            if k.startswith('classifier.') and 'experts' not in k and 'gate' not in k and k not in other_states:
                other_states[k] = []

    for k in other_states.keys():
        values = [state[k] for state in local_states if k in state]
        if values:
            other_states[k] = sum(client_weights[i] * values[i].float() for i in range(len(values)))

    final_state = {**shared_state, **expert_states, **gate_states, **other_states}
    return final_state


# ==== FL Training Loop ====
def run_fl_experiment(mode="fedavg", rounds=25, local_epochs=5, lr=1e-4,
                      client_loaders=None, test_loader=None, class_weights=None,
                      client_indices=None, class_names=None, num_experts=4):

    use_moe = (mode == "moe")
    model = get_inceptionresnetv2_model(num_classes=len(class_names), dropout_rate=0.4, use_moe=use_moe, num_experts=num_experts).to(device)

    comm_tracker = CommCostTracker(model)
    comm_details = comm_tracker.get_detailed_costs()

    print(f"\n{'='*80}")
    print(f"Starting {mode.upper()} Training with Inception-ResNet-V2")
    print(f"{'='*80}")
    print(f"Parameters (M): {comm_details['parameters_m']:.2f}")
    print(f"Uplink/Round (MB): {comm_details['uplink_mb']:.2f}")
    print(f"Downlink/Round (MB): {comm_details['downlink_mb']:.2f}")
    print(f"Total Comm./Round (MB): {comm_details['total_comm_mb']:.2f}")

    history = []
    comm_costs = []
    comp_times = []
    client_times = []
    best_f1 = 0
    patience_counter = 0
    best_model_state = None

    for round_num in range(rounds):
        round_start = time.time()
        local_states = []
        round_client_times = []

        for client_id, loader in enumerate(client_loaders):
            client_model = get_inceptionresnetv2_model(num_classes=len(class_names), dropout_rate=0.4, use_moe=use_moe, num_experts=num_experts).to(device)
            client_model.load_state_dict(model.state_dict())
            state, loss, client_time = train_local_inceptionresnetv2(client_model, loader, epochs=local_epochs, lr=lr, weight_decay=1e-4, class_weights=class_weights)
            local_states.append(state)
            round_client_times.append(client_time)
            del client_model
            gc.collect()
            torch.cuda.empty_cache()

        if mode == "fedavg" or mode == "dp_fedavg":
            client_weights = [len(idx) / sum(len(i) for i in client_indices) for idx in client_indices]
            aggregated_state = fedavg_enhanced(local_states, client_weights)
        elif mode == "fedprox":
            aggregated_state = fedprox_aggregation(local_states, model.state_dict())
        elif mode == "moe":
            client_weights = [len(idx) / sum(len(i) for i in client_indices) for idx in client_indices]
            aggregated_state = moe_aggregation(local_states, client_weights, num_experts)

        model.load_state_dict(aggregated_state)
        eval_results = eval_model_enhanced(model, test_loader, class_names)
        round_time = time.time() - round_start
        avg_client_time = np.mean(round_client_times)

        history.append(eval_results)
        comm_costs.append(comm_details['total_comm_mb'])
        comp_times.append(round_time)
        client_times.append(avg_client_time)

        log_str = (f"Round {round_num + 1:2d} | Acc: {eval_results['accuracy']:.3f} | "
                  f"F1: {eval_results['f1']:.3f} | AUC: {eval_results['auc_roc']:.3f} | "
                  f"Loss: {eval_results['loss']:.3f} | Client: {avg_client_time:.1f}s")
        if 'expert_entropy' in eval_results:
            log_str += f" | Entropy: {eval_results['expert_entropy']:.2f}"
        print(log_str)

        if eval_results['f1'] > best_f1:
            best_f1 = eval_results['f1']
            patience_counter = 0
            best_model_state = copy.deepcopy(model.state_dict())
        else:
            patience_counter += 1
            if patience_counter >= 7:
                print(f"Early stopping at round {round_num + 1}")
                if best_model_state is not None:
                    model.load_state_dict(best_model_state)
                break

    model.comm_details = comm_details
    model.client_times = client_times
    return model, history, comm_costs, comp_times


# ==== Visualization Functions ====
def plot_metrics_comparison(results_dict, class_names):
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    axes = axes.ravel()

    metrics = ['accuracy', 'f1', 'auc_roc', 'loss']
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']

    for i, metric in enumerate(metrics):
        for j, (method, (model, history, _, _)) in enumerate(results_dict.items()):
            values = [r[metric] for r in history]
            axes[i].plot(values, marker='o', label=method.upper(), linewidth=2, color=colors[j])

        axes[i].set_title(f'{metric.replace("_", " ").title()}', fontsize=12, fontweight='bold')
        axes[i].set_xlabel('Round')
        axes[i].set_ylabel(metric.replace("_", " ").title())
        axes[i].legend()
        axes[i].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('inceptionresnetv2_metrics_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()


def plot_roc_curves(best_results, class_names):
    plt.figure(figsize=(12, 8))
    y_true_bin = label_binarize(best_results['y_true'], classes=range(len(class_names)))
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#F39C12']

    for i, (class_name, color) in enumerate(zip(class_names, colors)):
        if len(np.unique(y_true_bin[:, i])) > 1:
            fpr, tpr, _ = roc_curve(y_true_bin[:, i], best_results['y_prob'][:, i])
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, color=color, linewidth=2, label=f'{class_name} (AUC = {roc_auc:.3f})')

    plt.plot([0, 1], [0, 1], 'k--', alpha=0.5)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('Multi-class ROC Curves - Inception-ResNet-V2', fontsize=14, fontweight='bold')
    plt.legend(loc="lower right")
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('inceptionresnetv2_roc_curves.png', dpi=300, bbox_inches='tight')
    plt.show()


def plot_confusion_matrix(y_true, y_pred, class_names):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names,
                yticklabels=class_names, cbar_kws={'label': 'Count'})
    plt.title('Confusion Matrix - Inception-ResNet-V2', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()
    plt.savefig('inceptionresnetv2_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()


def create_results_table(results_dict):
    fig, ax = plt.subplots(figsize=(18, 6))
    ax.axis('tight')
    ax.axis('off')

    headers = ['Method', 'Accuracy', 'F1-Score', 'AUC-ROC', 'Expert Entropy',
               'Parameters (M)', 'Uplink (MB)', 'Downlink (MB)', 'Total Comm (MB)',
               'Avg Client Time (s)', 'Round Time (s)']

    table_data = []
    for method, (model, history, comm_costs, comp_times) in results_dict.items():
        final = history[-1]
        comm = model.comm_details
        row = [
            method.upper(),
            f"{final['accuracy']:.3f}",
            f"{final['f1']:.3f}",
            f"{final['auc_roc']:.3f}",
            f"{final.get('expert_entropy', 0):.2f}",
            f"{comm['parameters_m']:.2f}",
            f"{comm['uplink_mb']:.2f}",
            f"{comm['downlink_mb']:.2f}",
            f"{comm['total_comm_mb']:.2f}",
            f"{np.mean(model.client_times):.1f}",
            f"{np.mean(comp_times):.1f}"
        ]
        table_data.append(row)

    table = ax.table(cellText=table_data, colLabels=headers, loc='center', cellLoc='center')
    table.auto_set_font_size(False)
    table.set_fontsize(9)
    table.scale(1.2, 2)

    for i in range(len(headers)):
        table[(0, i)].set_facecolor('#4ECDC4')
        table[(0, i)].set_text_props(weight='bold')

    for i in range(1, len(table_data) + 1):
        for j in range(len(headers)):
            if i % 2 == 0:
                table[(i, j)].set_facecolor('#F0F0F0')

    plt.title('Inception-ResNet-V2 Federated Learning - Complete Results', fontsize=14, fontweight='bold', pad=20)
    plt.savefig('inceptionresnetv2_results_table.png', dpi=300, bbox_inches='tight')
    plt.show()


def plot_communication_analysis(results_dict):
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))

    methods = list(results_dict.keys())
    params = [results_dict[m][0].comm_details['parameters_m'] for m in methods]
    uplink = [results_dict[m][0].comm_details['uplink_mb'] for m in methods]
    downlink = [results_dict[m][0].comm_details['downlink_mb'] for m in methods]
    total_comm = [results_dict[m][0].comm_details['total_comm_mb'] for m in methods]
    client_times = [np.mean(results_dict[m][0].client_times) for m in methods]
    round_times = [np.mean(results_dict[m][3]) for m in methods]
    final_f1 = [results_dict[m][1][-1]['f1'] for m in methods]

    axes[0, 0].bar([m.upper() for m in methods], params, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])
    axes[0, 0].set_title('Model Parameters (M)', fontweight='bold')
    axes[0, 0].set_ylabel('Parameters (Millions)')
    axes[0, 0].tick_params(axis='x', rotation=15)

    x = np.arange(len(methods))
    width = 0.25
    axes[0, 1].bar(x - width, uplink, width, label='Uplink', color='#FF6B6B')
    axes[0, 1].bar(x, downlink, width, label='Downlink', color='#4ECDC4')
    axes[0, 1].bar(x + width, total_comm, width, label='Total', color='#45B7D1')
    axes[0, 1].set_title('Communication Costs per Round (MB)', fontweight='bold')
    axes[0, 1].set_xticks(x)
    axes[0, 1].set_xticklabels([m.upper() for m in methods], rotation=15)
    axes[0, 1].legend()

    axes[1, 0].bar(x - width/2, client_times, width, label='Client Time', color='#96CEB4')
    axes[1, 0].bar(x + width/2, round_times, width, label='Round Time', color='#F39C12')
    axes[1, 0].set_title('Training Times (seconds)', fontweight='bold')
    axes[1, 0].set_xticks(x)
    axes[1, 0].set_xticklabels([m.upper() for m in methods], rotation=15)
    axes[1, 0].legend()

    efficiency = [f1 / comm if comm > 0 else 0 for f1, comm in zip(final_f1, total_comm)]
    axes[1, 1].bar([m.upper() for m in methods], efficiency, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])
    axes[1, 1].set_title('Communication Efficiency (F1/MB)', fontweight='bold')
    axes[1, 1].set_ylabel('F1 Score per MB')
    axes[1, 1].tick_params(axis='x', rotation=15)

    plt.tight_layout()
    plt.savefig('inceptionresnetv2_communication_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()


# ==== Main Experiment Runner ====
def run_complete_experiment(base_path):
    print("="*80)
    print("INCEPTION-RESNET-V2 FEDERATED LEARNING - APTOS DATASET")
    print("="*80)

    print("\nSetting up APTOS dataset...")
    train_ds, test_ds, class_names, class_weights, num_classes = setup_aptos_dataset_with_split(base_path)
    test_loader = DataLoader(test_ds, batch_size=16, shuffle=False, num_workers=2)

    labels = np.array([y for _, y in train_ds.samples])
    client_indices = split_noniid_aptos(labels, n_clients=4, alpha=0.5)

    print("\nClient Data Distribution:")
    for i, idx in enumerate(client_indices):
        client_labels = labels[idx]
        unique, counts = np.unique(client_labels, return_counts=True)
        class_dist = {class_names[u]: c for u, c in zip(unique, counts)}
        print(f"  Client {i+1}: {len(idx)} samples - {class_dist}")

    client_loaders = []
    for idx in client_indices:
        client_labels = labels[idx]
        sample_weights = class_weights[client_labels]
        sampler = WeightedRandomSampler(sample_weights, len(sample_weights), replacement=True)
        loader = DataLoader(Subset(train_ds, idx), batch_size=12, sampler=sampler, num_workers=2)
        client_loaders.append(loader)

    results = {}
    methods = ['fedavg', 'fedprox', 'moe']

    for method in methods:
        print(f"\n{'='*80}")
        print(f"EXPERIMENT: {method.upper()}")
        print(f"{'='*80}")

        model, history, comm, comp = run_fl_experiment(
            mode=method,
            rounds=25,
            local_epochs=5,
            lr=1e-4,
            client_loaders=client_loaders,
            test_loader=test_loader,
            class_weights=class_weights,
            client_indices=client_indices,
            class_names=class_names,
            num_experts=4
        )
        results[method] = (model, history, comm, comp)

        final = history[-1]
        print(f"\n{method.upper()} Final Results:")
        print(f"  Accuracy: {final['accuracy']:.3f}")
        print(f"  F1-Score: {final['f1']:.3f}")
        print(f"  AUC-ROC: {final['auc_roc']:.3f}")
        if 'expert_entropy' in final:
            print(f"  Expert Entropy: {final['expert_entropy']:.2f}")

    print(f"\n{'='*80}")
    print(f"EXPERIMENT: DP-FEDAVG")
    print(f"{'='*80}")

    model_dp, history_dp, comm_dp, comp_dp = run_fl_experiment(
        mode='dp_fedavg',
        rounds=25,
        local_epochs=5,
        lr=1e-4,
        client_loaders=client_loaders,
        test_loader=test_loader,
        class_weights=class_weights,
        client_indices=client_indices,
        class_names=class_names
    )
    results['dp_fedavg'] = (model_dp, history_dp, comm_dp, comp_dp)

    print(f"\n{'='*80}")
    print("GENERATING VISUALIZATIONS")
    print(f"{'='*80}")

    print("\n1. Plotting metrics...")
    plot_metrics_comparison(results, class_names)

    print("2. Creating results table...")
    create_results_table(results)

    print("3. Communication analysis...")
    plot_communication_analysis(results)

    best_method = max(results.keys(), key=lambda x: results[x][1][-1]['f1'])
    best_model, best_history = results[best_method][:2]
    best_results = best_history[-1]

    print(f"\n{'='*80}")
    print(f"BEST METHOD: {best_method.upper()}")
    print(f"{'='*80}")
    print(f"\nFinal Metrics:")
    print(f"  Accuracy:  {best_results['accuracy']:.3f}")
    print(f"  F1-Score:  {best_results['f1']:.3f}")
    print(f"  AUC-ROC:   {best_results['auc_roc']:.3f}")
    if 'expert_entropy' in best_results:
        print(f"  Expert Entropy: {best_results['expert_entropy']:.2f}")

    print("\n4. ROC curves...")
    plot_roc_curves(best_results, class_names)

    print("5. Confusion matrix...")
    plot_confusion_matrix(best_results['y_true'], best_results['y_pred'], class_names)

    print(f"\n{'='*80}")
    print(f"CLASSIFICATION REPORT - {best_method.upper()}")
    print(f"{'='*80}")
    print(classification_report(best_results['y_true'], best_results['y_pred'], target_names=class_names, digits=3))

    print(f"\n{'='*80}")
    print("EXPERIMENT COMPLETED!")
    print(f"{'='*80}")
    print("\nGenerated files:")
    print("  - inceptionresnetv2_metrics_comparison.png")
    print("  - inceptionresnetv2_results_table.png")
    print("  - inceptionresnetv2_communication_analysis.png")
    print("  - inceptionresnetv2_roc_curves.png")
    print("  - inceptionresnetv2_confusion_matrix.png")

    return results, best_model, best_method, class_names


# ==== RUN EXPERIMENT ====
if __name__ == "__main__":
    BASE_PATH = "/content/drive/MyDrive/Retinopathy/APTOS/gaussian_filtered_images/gaussian_filtered_images"

    print("\n" + "="*80)
    print("INCEPTION-RESNET-V2 FEDERATED LEARNING")
    print("="*80)
    print("\nArchitecture Features:")
    print("  • Combines Inception modules with Residual connections")
    print("  • Multi-scale feature extraction")
    print("  • ~55M parameters (deep & powerful)")
    print("  • 299x299 input resolution")
    print("  • State-of-the-art for image classification")
    print("\nRunning 4 FL methods with complete metrics...\n")

    results, best_model, best_method, class_names = run_complete_experiment(BASE_PATH)

    print(f"\nBest method: {best_method.upper()}")
    print("All visualizations saved!")
```
