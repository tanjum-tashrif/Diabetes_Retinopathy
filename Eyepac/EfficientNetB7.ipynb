```python
import os, time, copy, gc
import numpy as np, matplotlib.pyplot as plt, seaborn as sns
import torch, torch.nn as nn, torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, Subset, WeightedRandomSampler
from sklearn.metrics import accuracy_score, f1_score, recall_score, confusion_matrix, roc_curve, precision_recall_curve, auc, classification_report
from sklearn.preprocessing import label_binarize
from torchcam.methods import SmoothGradCAMpp
from torchcam.utils import overlay_mask
import warnings
warnings.filterwarnings('ignore')

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python


# ==== APTOS Dataset Setup with Train/Test Split ====
def setup_aptos_dataset_with_split(base_path, test_ratio=0.2, random_seed=42):
    """Setup APTOS dataset and create train/test split for each class"""

    np.random.seed(random_seed)
    torch.manual_seed(random_seed)

    # EfficientNet-specific transforms
    train_tfm = transforms.Compose([
        transforms.Resize((456, 456)),  # Larger initial size for EfficientNetB7
        transforms.RandomResizedCrop(224, scale=(0.85, 1.0)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomVerticalFlip(p=0.5),
        transforms.RandomRotation(15),
        transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.1),
        transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    test_tfm = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    class_folders = [f for f in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, f))]
    class_folders.sort()

    print(f"Found classes: {class_folders}")

    all_samples = []
    class_to_idx = {class_name: idx for idx, class_name in enumerate(class_folders)}

    for class_name in class_folders:
        class_path = os.path.join(base_path, class_name)
        class_idx = class_to_idx[class_name]

        image_files = [f for f in os.listdir(class_path)
                      if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))]

        print(f"Class '{class_name}': {len(image_files)} samples")

        for img_file in image_files:
            img_path = os.path.join(class_path, img_file)
            all_samples.append((img_path, class_idx))

    print(f"Total samples: {len(all_samples)}")

    train_samples = []
    test_samples = []
    class_counts = torch.zeros(len(class_folders))

    for class_idx, class_name in enumerate(class_folders):
        class_samples = [(path, label) for path, label in all_samples if label == class_idx]
        np.random.shuffle(class_samples)

        n_test = max(1, int(len(class_samples) * test_ratio))
        n_train = len(class_samples) - n_test

        train_samples.extend(class_samples[:n_train])
        test_samples.extend(class_samples[n_train:])

        class_counts[class_idx] = n_train
        print(f"  {class_name}: {n_train} train, {n_test} test")

    np.random.shuffle(train_samples)
    np.random.shuffle(test_samples)

    class APTOSDataset(torch.utils.data.Dataset):
        def __init__(self, samples, transform=None):
            self.samples = samples
            self.transform = transform

        def __len__(self):
            return len(self.samples)

        def __getitem__(self, idx):
            path, label = self.samples[idx]
            from PIL import Image
            image = Image.open(path).convert('RGB')

            if self.transform:
                image = self.transform(image)

            return image, label

    train_ds = APTOSDataset(train_samples, transform=train_tfm)
    test_ds = APTOSDataset(test_samples, transform=test_tfm)

    train_ds.classes = class_folders
    test_ds.classes = class_folders
    train_ds.samples = train_samples
    test_ds.samples = test_samples

    num_classes = len(class_folders)

    print(f"\nAPTOS Dataset split completed:")
    print(f"Classes: {class_folders}")
    print(f"Number of classes: {num_classes}")
    print(f"Train samples: {len(train_ds)}")
    print(f"Test samples: {len(test_ds)}")

    # Calculate class weights
    class_weights = 1.0 / (class_counts + 1e-8)
    class_weights = class_weights / class_weights.sum() * num_classes

    print(f"\nClass distribution (train):")
    for i, (name, count) in enumerate(zip(class_folders, class_counts.tolist())):
        print(f"  {name}: {int(count)} samples (weight: {class_weights[i]:.3f})")

    return train_ds, test_ds, class_folders, class_weights, num_classes

# ==== Enhanced Non-IID Data Splitting ====
def split_noniid_aptos(labels, n_clients=4, alpha=0.3, min_samples_per_class=5):
    """Enhanced non-IID data splitting for APTOS"""
    idxs = np.arange(len(labels))
    client_idx = [[] for _ in range(n_clients)]

    unique_classes = np.unique(labels)
    print(f"Splitting data across {n_clients} clients with alpha={alpha}")

    for c in unique_classes:
        idx_c = idxs[labels == c]
        np.random.shuffle(idx_c)

        min_per_client = min(min_samples_per_class, len(idx_c) // n_clients)

        if len(idx_c) < n_clients * min_per_client:
            splits = np.array_split(idx_c, n_clients)
            print(f"  Class {c}: Even split ({len(idx_c)} samples)")
        else:
            props = np.random.dirichlet([alpha] * n_clients)
            props = np.maximum(props, min_per_client / len(idx_c))
            props = props / props.sum()

            splits = (props.cumsum()[:-1] * len(idx_c)).astype(int)
            splits = np.split(idx_c, splits)
            print(f"  Class {c}: Dirichlet split ({len(idx_c)} samples)")

        for i in range(n_clients):
            client_idx[i] += splits[i].tolist()

    for i in range(n_clients):
        np.random.shuffle(client_idx[i])

    return client_idx

# ==== Enhanced EfficientNetB7 Model ====
def get_efficientnet_model(num_classes=5, pretrained=True, dropout_rate=0.4, use_moe=False, num_experts=4):
    """Enhanced EfficientNetB7 model for APTOS"""

    # Load pretrained EfficientNetB7
    model = models.efficientnet_b7(pretrained=pretrained)

    # Get the number of input features for the classifier
    num_features = model.classifier[1].in_features

    if use_moe:
        # Create MOE-based classifier
        model.classifier = nn.Sequential(
            nn.Dropout(dropout_rate),
            MixtureOfExperts(
                input_dim=num_features,
                hidden_dim=num_features // 2,
                output_dim=num_classes,
                num_experts=num_experts,
                dropout_rate=dropout_rate,
                top_k=2
            )
        )
        model.is_moe = True
    else:
        # Create enhanced standard classifier with more capacity
        model.classifier = nn.Sequential(
            nn.Dropout(dropout_rate),
            nn.Linear(num_features, num_features // 2),
            nn.ReLU(inplace=True),
            nn.BatchNorm1d(num_features // 2),
            nn.Dropout(dropout_rate),
            nn.Linear(num_features // 2, num_features // 4),
            nn.ReLU(inplace=True),
            nn.BatchNorm1d(num_features // 4),
            nn.Dropout(dropout_rate // 2),
            nn.Linear(num_features // 4, num_classes)
        )
        model.is_moe = False

    # Initialize weights
    if hasattr(model.classifier, 'experts') or any(hasattr(m, 'experts') for m in model.classifier.modules()):
        # Initialize MOE experts
        for m in model.classifier.modules():
            if hasattr(m, 'experts'):
                for expert in m.experts:
                    for layer in expert.modules():
                        if isinstance(layer, nn.Linear):
                            nn.init.kaiming_normal_(layer.weight, mode='fan_out', nonlinearity='relu')
                            if layer.bias is not None:
                                nn.init.constant_(layer.bias, 0)
    else:
        # Initialize standard classifier
        for m in model.classifier.modules():
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)

    return model

# ==== MOE Components ====
class ExpertNetwork(nn.Module):
    """Individual expert network for APTOS classification"""
    def __init__(self, input_dim, hidden_dim, output_dim, dropout_rate=0.3):
        super(ExpertNetwork, self).__init__()
        self.network = nn.Sequential(
            nn.LayerNorm(input_dim),
            nn.Linear(input_dim, hidden_dim),
            nn.GELU(),
            nn.Dropout(dropout_rate),
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.GELU(),
            nn.Dropout(dropout_rate),
            nn.Linear(hidden_dim // 2, output_dim)
        )

    def forward(self, x):
        return self.network(x)

class GatingNetwork(nn.Module):
    """Gating network for MOE"""
    def __init__(self, input_dim, num_experts, dropout_rate=0.1):
        super(GatingNetwork, self).__init__()
        self.gate = nn.Sequential(
            nn.LayerNorm(input_dim),
            nn.Linear(input_dim, input_dim // 2),
            nn.GELU(),
            nn.Dropout(dropout_rate),
            nn.Linear(input_dim // 2, num_experts),
            nn.Softmax(dim=-1)
        )

    def forward(self, x):
        return self.gate(x)

class MixtureOfExperts(nn.Module):
    """Enhanced Mixture of Experts for APTOS"""
    def __init__(self, input_dim, hidden_dim, output_dim, num_experts=4, dropout_rate=0.3, top_k=2):
        super(MixtureOfExperts, self).__init__()
        self.num_experts = num_experts
        self.top_k = top_k

        self.experts = nn.ModuleList([
            ExpertNetwork(input_dim, hidden_dim, output_dim, dropout_rate)
            for _ in range(num_experts)
        ])

        self.gate = GatingNetwork(input_dim, num_experts, dropout_rate)

    def forward(self, x):
        gate_weights = self.gate(x)

        top_k_weights, top_k_indices = torch.topk(gate_weights, self.top_k, dim=-1)
        top_k_weights = top_k_weights / top_k_weights.sum(dim=-1, keepdim=True)

        expert_outputs = []
        for i, expert in enumerate(self.experts):
            expert_outputs.append(expert(x))
        expert_outputs = torch.stack(expert_outputs, dim=1)

        batch_size, output_dim = x.size(0), expert_outputs.size(-1)
        final_output = torch.zeros(batch_size, output_dim, device=x.device)

        for i in range(self.top_k):
            expert_idx = top_k_indices[:, i]
            expert_weight = top_k_weights[:, i].unsqueeze(-1)
            selected_outputs = expert_outputs[torch.arange(batch_size), expert_idx]
            final_output += expert_weight * selected_outputs

        return final_output, gate_weights

# ==== Loss Functions ====
class FocalLoss(nn.Module):
    """Focal Loss for imbalanced APTOS dataset"""
    def __init__(self, alpha=None, gamma=2.0, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        ce_loss = nn.functional.cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = (1 - pt) ** self.gamma * ce_loss

        if self.alpha is not None:
            alpha_t = self.alpha[targets]
            focal_loss = alpha_t * focal_loss

        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss

class MOELoss(nn.Module):
    """MOE Loss with load balancing"""
    def __init__(self, base_criterion, load_balance_weight=0.01):
        super(MOELoss, self).__init__()
        self.base_criterion = base_criterion
        self.load_balance_weight = load_balance_weight

    def forward(self, outputs, targets):
        if isinstance(outputs, tuple):
            logits, gate_weights = outputs
            base_loss = self.base_criterion(logits, targets)
        else:
            logits = outputs
            base_loss = self.base_criterion(logits, targets)
            gate_weights = None

        total_loss = base_loss

        if gate_weights is not None and self.load_balance_weight > 0:
            num_experts = gate_weights.size(-1)
            mean_gate_weights = gate_weights.mean(dim=0)
            uniform_target = torch.ones_like(mean_gate_weights) / num_experts

            load_balance_loss = nn.functional.kl_div(
                torch.log(mean_gate_weights + 1e-8),
                uniform_target,
                reduction='batchmean'
            )

            total_loss += self.load_balance_weight * load_balance_loss

        return total_loss

# ==== Communication Cost Tracker ====
class CommCostTracker:
    """Track detailed communication costs"""
    def __init__(self, model):
        self.model = model
        self.param_count = self.count_parameters()
        self.param_size_mb = self.param_count * 4 / (1024 ** 2)  # 4 bytes per float32

    def count_parameters(self):
        return sum(p.numel() for p in self.model.parameters())

    def get_uplink_cost(self):
        """Cost for client to send model to server (MB)"""
        return self.param_size_mb

    def get_downlink_cost(self):
        """Cost for server to send model to client (MB)"""
        return self.param_size_mb

    def get_total_comm_cost(self):
        """Total communication cost per round (MB)"""
        return self.param_size_mb * 2  # Uplink + Downlink

    def get_detailed_costs(self):
        """Get all communication costs"""
        return {
            'parameters_m': self.param_count / 1e6,
            'uplink_mb': self.get_uplink_cost(),
            'downlink_mb': self.get_downlink_cost(),
            'total_comm_mb': self.get_total_comm_cost()
        }

# ==== Enhanced Training Functions ====
def train_local_enhanced_efficientnet(model, loader, epochs=5, lr=1e-3, weight_decay=1e-4, class_weights=None):
    """Enhanced local training for EfficientNetB7"""
    model = model.to(device)
    model.train()

    start_time = time.time()

    # Different learning rates for pretrained features and new classifier
    feature_params = []
    classifier_params = []

    for name, param in model.named_parameters():
        if 'classifier' in name:
            classifier_params.append(param)
        else:
            feature_params.append(param)

    # EfficientNet needs careful learning rate tuning
    optimizer = optim.AdamW([
        {'params': feature_params, 'lr': lr * 0.1},  # Lower LR for pretrained features
        {'params': classifier_params, 'lr': lr}       # Higher LR for classifier
    ], weight_decay=weight_decay)

    base_criterion = FocalLoss(alpha=class_weights.to(device) if class_weights is not None else None, gamma=2.0)
    criterion = MOELoss(base_criterion, load_balance_weight=0.01 if hasattr(model, 'is_moe') and model.is_moe else 0.0)

    # Cosine annealing with warmup
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=lr*0.01)

    total_loss = 0
    total_samples = 0

    for epoch in range(epochs):
        epoch_loss = 0
        epoch_samples = 0

        for batch_idx, (X, y) in enumerate(loader):
            X, y = X.to(device), y.to(device)

            optimizer.zero_grad()
            outputs = model(X)
            loss = criterion(outputs, y)
            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            optimizer.step()

            epoch_loss += loss.item() * X.size(0)
            epoch_samples += X.size(0)

        scheduler.step()
        total_loss += epoch_loss
        total_samples += epoch_samples

    training_time = time.time() - start_time
    avg_loss = total_loss / total_samples if total_samples > 0 else 0

    return copy.deepcopy(model.state_dict()), avg_loss, training_time

# ==== Enhanced Evaluation ====
def eval_model_enhanced_aptos(model, loader, class_names):
    """Enhanced evaluation for APTOS"""
    model.eval()
    y_true, y_pred, y_prob = [], [], []
    total_loss = 0
    gate_weights_all = []
    criterion = nn.CrossEntropyLoss()

    with torch.no_grad():
        for X, y in loader:
            X, y = X.to(device), y.to(device)
            outputs = model(X)

            if hasattr(model, 'is_moe') and model.is_moe and isinstance(outputs, tuple):
                logits, gate_weights = outputs
                gate_weights_all.append(gate_weights.cpu())
            else:
                logits = outputs

            loss = criterion(logits, y)
            total_loss += loss.item() * X.size(0)

            probs = torch.softmax(logits, dim=1)
            preds = logits.argmax(dim=1)

            y_true.extend(y.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())
            y_prob.extend(probs.cpu().numpy())

    y_true, y_pred, y_prob = np.array(y_true), np.array(y_pred), np.array(y_prob)

    accuracy = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average='weighted')
    recall = recall_score(y_true, y_pred, average='weighted')

    # Multi-class AUC
    try:
        y_true_bin = label_binarize(y_true, classes=range(len(class_names)))
        auc_scores = []
        for i in range(len(class_names)):
            if len(np.unique(y_true_bin[:, i])) > 1:
                fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_prob[:, i])
                auc_scores.append(auc(fpr, tpr))
            else:
                auc_scores.append(0.0)
        auc_roc = np.mean(auc_scores)
    except:
        auc_roc = 0

    avg_loss = total_loss / len(loader.dataset)

    results = {
        'accuracy': accuracy,
        'f1': f1,
        'recall': recall,
        'auc_roc': auc_roc,
        'loss': avg_loss,
        'y_true': y_true,
        'y_pred': y_pred,
        'y_prob': y_prob,
        'per_class_auc': auc_scores if 'auc_scores' in locals() else [0] * len(class_names)
    }

    if gate_weights_all:
        gate_weights_all = torch.cat(gate_weights_all, dim=0)
        expert_usage = gate_weights_all.mean(dim=0).numpy()
        expert_entropy = -np.sum(expert_usage * np.log(expert_usage + 1e-8))
        results['expert_usage'] = expert_usage
        results['expert_entropy'] = expert_entropy

    return results

# ==== Enhanced Grad-CAM Implementation ====
def generate_gradcam_visualization(model, test_loader, class_names, num_samples=5):
    """Generate Grad-CAM visualizations for APTOS samples"""
    model.eval()

    # Get the last convolutional layer for EfficientNet
    target_layer = None
    for name, module in model.named_modules():
        if isinstance(module, nn.Conv2d):
            target_layer = module

    if target_layer is None:
        print("Warning: Could not find Conv2d layer for Grad-CAM")
        return

    try:
        cam_extractor = SmoothGradCAMpp(model, target_layer=target_layer)

        # Get some test samples
        dataiter = iter(test_loader)
        images, labels = next(dataiter)

        # Select first few samples
        images = images[:num_samples].to(device)
        labels = labels[:num_samples]

        with torch.no_grad():
            outputs = model(images)
            if isinstance(outputs, tuple):
                outputs = outputs[0]  # For MOE models
            predictions = outputs.argmax(dim=1)

        fig, axes = plt.subplots(2, num_samples, figsize=(15, 6))
        if num_samples == 1:
            axes = axes.reshape(2, 1)

        for i in range(num_samples):
            # Original image
            img = images[i].cpu()
            img_denorm = img * torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1) + torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
            img_denorm = torch.clamp(img_denorm, 0, 1)

            # Generate CAM
            cam = cam_extractor(predictions[i].item(), outputs[i:i+1])

            if len(cam) > 0:
                activation_map = cam[0].squeeze().cpu()

                # Show original image
                axes[0, i].imshow(img_denorm.permute(1, 2, 0))
                axes[0, i].set_title(f'Original\nTrue: {class_names[labels[i]]}\nPred: {class_names[predictions[i]]}')
                axes[0, i].axis('off')

                # Show Grad-CAM overlay
                result = overlay_mask(img_denorm.permute(1, 2, 0), activation_map, alpha=0.5)
                axes[1, i].imshow(result)
                axes[1, i].set_title('Grad-CAM Overlay')
                axes[1, i].axis('off')
            else:
                print(f"Warning: Could not generate CAM for sample {i}")

        plt.tight_layout()
        plt.suptitle('Grad-CAM Visualizations for APTOS Dataset', fontsize=16, y=1.02)
        plt.show()

    except Exception as e:
        print(f"Error generating Grad-CAM: {str(e)}")
        print("This might be due to model architecture compatibility issues")

# ==== Enhanced FL Training Loop ====
def run_enhanced_fl_aptos(mode="fedavg", rounds=15, local_epochs=3, dp=False, sigma=0.01,
                         lr=1e-3, patience=5, client_loaders=None, test_loader=None,
                         class_weights=None, client_indices=None, class_names=None, num_experts=4):
    """Enhanced federated learning with detailed metrics tracking"""

    use_moe = (mode == "moe")
    model = get_efficientnet_model(
        num_classes=len(class_names),
        use_moe=use_moe,
        num_experts=num_experts
    ).to(device)

    # Initialize communication cost tracker
    comm_tracker = CommCostTracker(model)
    comm_details = comm_tracker.get_detailed_costs()

    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"Model: EfficientNetB7 {'with MOE' if use_moe else ''}")
    print(f"Total parameters: {total_params:,}")
    print(f"Trainable parameters: {trainable_params:,}")
    print(f"Parameters (M): {comm_details['parameters_m']:.2f}")
    print(f"Uplink/Round (MB): {comm_details['uplink_mb']:.2f}")
    print(f"Downlink/Round (MB): {comm_details['downlink_mb']:.2f}")
    print(f"Total Comm./Round (MB): {comm_details['total_comm_mb']:.2f}")

    history = []
    comm_costs = []
    comp_times = []
    client_times = []
    best_f1 = 0
    patience_counter = 0
    best_model_state = None

    print(f"Starting {mode} FL training for APTOS with EfficientNetB7 for {rounds} rounds...")
    print(f"Local epochs: {local_epochs}, LR: {lr}, DP: {dp}")
    print("-" * 100)

    for round_num in range(rounds):
        round_start = time.time()
        local_states = []
        local_losses = []
        round_client_times = []

        # Local training on each client
        for client_id, loader in enumerate(client_loaders):
            client_model = get_efficientnet_model(
                num_classes=len(class_names),
                use_moe=use_moe,
                num_experts=num_experts
            ).to(device)
            client_model.load_state_dict(model.state_dict())

            # Train locally and track time
            state, loss, client_time = train_local_enhanced_efficientnet(
                client_model, loader, epochs=local_epochs,
                lr=lr, class_weights=class_weights
            )

            local_states.append(state)
            local_losses.append(loss)
            round_client_times.append(client_time)

            # Clean up
            del client_model
            gc.collect()
            torch.cuda.empty_cache()

        # Server aggregation
        if mode == "fedavg":
            client_weights = [len(idx) for idx in client_indices]
            total_samples = sum(client_weights)
            client_weights = [w / total_samples for w in client_weights]
            aggregated_state = fedavg_enhanced(local_states, client_weights)

        elif mode == "fedprox":
            aggregated_state = fedprox_aggregation(local_states, model.state_dict())

        elif mode == "moe":
            client_weights = [len(idx) for idx in client_indices]
            total_samples = sum(client_weights)
            client_weights = [w / total_samples for w in client_weights]
            aggregated_state = moe_aggregation(local_states, client_weights, num_experts)

        model.load_state_dict(aggregated_state)

        # Evaluation
        eval_results = eval_model_enhanced_aptos(model, test_loader, class_names)

        # Calculate metrics
        round_time = time.time() - round_start
        avg_client_time = np.mean(round_client_times)

        history.append(eval_results)
        comm_costs.append(comm_details['total_comm_mb'])
        comp_times.append(round_time)
        client_times.append(avg_client_time)

        # Enhanced logging with detailed metrics
        log_str = (f"Round {round_num + 1:2d} | "
                  f"Acc: {eval_results['accuracy']:.3f} | "
                  f"F1: {eval_results['f1']:.3f} | "
                  f"AUC: {eval_results['auc_roc']:.3f} | "
                  f"Loss: {eval_results['loss']:.3f} | "
                  f"Avg Client Time: {avg_client_time:.1f}s | "
                  f"Round Time: {round_time:.1f}s")

        if 'expert_entropy' in eval_results:
            log_str += f" | Expert Entropy: {eval_results['expert_entropy']:.2f}"

        print(log_str)

        # Early stopping
        if eval_results['f1'] > best_f1:
            best_f1 = eval_results['f1']
            patience_counter = 0
            best_model_state = copy.deepcopy(model.state_dict())
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f"Early stopping at round {round_num + 1}")
                if best_model_state is not None:
                    model.load_state_dict(best_model_state)
                break

    # Store communication details in the model for later access
    model.comm_details = comm_details
    model.client_times = client_times

    return model, history, comm_costs, comp_times

# ==== Aggregation Methods ====
def fedavg_enhanced(states, weights=None):
    """Enhanced FedAvg aggregation"""
    if weights is None:
        weights = [1.0 / len(states)] * len(states)

    avg_state = {}
    for key in states[0].keys():
        if states[0][key].dtype not in [torch.float32, torch.float64, torch.float16]:
            avg_state[key] = states[0][key].clone()
        else:
            avg_state[key] = torch.zeros_like(states[0][key], dtype=torch.float32)
            for i, state in enumerate(states):
                avg_state[key] += weights[i] * state[key].float()
    return avg_state

def fedprox_aggregation(states, global_state, mu=0.01):
    """FedProx-style aggregation with proximal term"""
    avg_state = fedavg_enhanced(states)
    for key in avg_state.keys():
        if key in global_state and avg_state[key].dtype in [torch.float32, torch.float64, torch.float16]:
            avg_state[key] = (1 - mu) * avg_state[key] + mu * global_state[key].float()
    return avg_state

def moe_aggregation(local_states, client_weights=None, num_experts=4):
    """MOE-specific aggregation"""
    if client_weights is None:
        client_weights = [1.0 / len(local_states)] * len(local_states)

    # Aggregate shared backbone (features)
    shared_state = fedavg_enhanced([
        {k: v for k, v in state.items() if 'classifier' not in k}
        for state in local_states
    ], client_weights)

    # Aggregate each expert separately
    expert_states = {}
    for expert_idx in range(num_experts):
        expert_key_prefix = f'classifier.1.experts.{expert_idx}'
        expert_params = [
            {k.replace(expert_key_prefix + '.', ''): v for k, v in state.items()
             if k.startswith(expert_key_prefix)}
            for state in local_states
        ]

        expert_params = [ep for ep in expert_params if ep]
        if expert_params:
            aggregated_expert = fedavg_enhanced(expert_params, client_weights)
            for k, v in aggregated_expert.items():
                expert_states[f'{expert_key_prefix}.{k}'] = v

    # Aggregate gating network
    gate_params = [
        {k.replace('classifier.1.gate.', ''): v for k, v in state.items()
         if k.startswith('classifier.1.gate.')}
        for state in local_states
    ]

    gate_states = {}
    if gate_params and gate_params[0]:
        aggregated_gate = fedavg_enhanced(gate_params, client_weights)
        gate_states = {f'classifier.1.gate.{k}': v for k, v in aggregated_gate.items()}

    # Aggregate other classifier components (like dropout)
    other_classifier_states = {}
    for state in local_states:
        for k, v in state.items():
            if k.startswith('classifier.') and 'experts' not in k and 'gate' not in k:
                if k not in other_classifier_states:
                    other_classifier_states[k] = []
                other_classifier_states[k].append(v)

    for k, v_list in other_classifier_states.items():
        if len(v_list) > 0:
            weighted_sum = torch.zeros_like(v_list[0], dtype=torch.float32)
            for i, v in enumerate(v_list):
                weighted_sum += client_weights[i] * v.float()
            expert_states[k] = weighted_sum

    # Combine all states
    final_state = {}
    final_state.update(shared_state)
    final_state.update(expert_states)
    final_state.update(gate_states)

    return final_state

# ==== Enhanced Visualization Functions ====
def plot_enhanced_results_aptos(histories, labels, class_names, save_plots=True):
    """Enhanced plotting for APTOS results"""
    metrics = ['accuracy', 'f1', 'auc_roc', 'loss']
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    axes = axes.ravel()

    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#F39C12']

    for i, metric in enumerate(metrics):
        for j, (hist, label) in enumerate(zip(histories, labels)):
            values = [round_data[metric] for round_data in hist]
            axes[i].plot(values, marker='o', label=label, linewidth=2,
                        markersize=4, color=colors[j % len(colors)])

        axes[i].set_title(f'{metric.replace("_", " ").title()} vs Round',
                         fontsize=12, fontweight='bold')
        axes[i].set_xlabel('Round')
        axes[i].set_ylabel(metric.replace("_", " ").title())
        axes[i].legend()
        axes[i].grid(True, alpha=0.3)

        # Set appropriate y-limits
        if metric == 'loss':
            axes[i].set_ylim(bottom=0)
        else:
            axes[i].set_ylim(0, 1)

    plt.tight_layout()
    if save_plots:
        plt.savefig('aptos_efficientnet_metrics.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_communication_analysis(models_results, labels):
    """Plot detailed communication analysis"""
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))

    # Extract communication details
    comm_data = []
    for model, history, comm_costs, comp_times in models_results:
        if hasattr(model, 'comm_details'):
            details = model.comm_details.copy()
            details['avg_client_time'] = np.mean(model.client_times) if hasattr(model, 'client_times') else 0
            details['avg_round_time'] = np.mean(comp_times)
            comm_data.append(details)
        else:
            # Fallback if comm_details not available
            comm_tracker = CommCostTracker(model)
            details = comm_tracker.get_detailed_costs()
            details['avg_client_time'] = np.mean(comp_times) / 4  # Estimate
            details['avg_round_time'] = np.mean(comp_times)
            comm_data.append(details)

    # Parameters comparison
    params = [data['parameters_m'] for data in comm_data]
    axes[0,0].bar(labels, params, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'][:len(labels)])
    axes[0,0].set_title('Model Parameters (M)', fontweight='bold')
    axes[0,0].set_ylabel('Parameters (Millions)')
    axes[0,0].tick_params(axis='x', rotation=45)

    # Communication costs
    uplink = [data['uplink_mb'] for data in comm_data]
    downlink = [data['downlink_mb'] for data in comm_data]
    total_comm = [data['total_comm_mb'] for data in comm_data]

    x = np.arange(len(labels))
    width = 0.25

    axes[0,1].bar(x - width, uplink, width, label='Uplink', alpha=0.8, color='#FF6B6B')
    axes[0,1].bar(x, downlink, width, label='Downlink', alpha=0.8, color='#4ECDC4')
    axes[0,1].bar(x + width, total_comm, width, label='Total', alpha=0.8, color='#45B7D1')

    axes[0,1].set_title('Communication Costs per Round (MB)', fontweight='bold')
    axes[0,1].set_ylabel('MB')
    axes[0,1].set_xticks(x)
    axes[0,1].set_xticklabels(labels, rotation=45)
    axes[0,1].legend()

    # Timing analysis
    client_times = [data['avg_client_time'] for data in comm_data]
    round_times = [data['avg_round_time'] for data in comm_data]

    x = np.arange(len(labels))
    width = 0.35

    axes[1,0].bar(x - width/2, client_times, width, label='Avg Client Time', alpha=0.8, color='#96CEB4')
    axes[1,0].bar(x + width/2, round_times, width, label='Round Time', alpha=0.8, color='#F39C12')

    axes[1,0].set_title('Training Times (seconds)', fontweight='bold')
    axes[1,0].set_ylabel('Time (s)')
    axes[1,0].set_xticks(x)
    axes[1,0].set_xticklabels(labels, rotation=45)
    axes[1,0].legend()

    # Communication efficiency (Performance per MB)
    # Use final F1 scores
    final_f1s = []
    for _, history, _, _ in models_results:
        final_f1s.append(history[-1]['f1'])

    efficiency = [f1 / comm for f1, comm in zip(final_f1s, total_comm)]

    axes[1,1].bar(labels, efficiency, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'][:len(labels)])
    axes[1,1].set_title('Communication Efficiency (F1/MB)', fontweight='bold')
    axes[1,1].set_ylabel('F1 Score per MB')
    axes[1,1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.savefig('aptos_communication_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_performance_table(models_results, labels, class_names):
    """Create a detailed performance table"""
    fig, ax = plt.subplots(figsize=(14, 8))
    ax.axis('tight')
    ax.axis('off')

    # Prepare data
    table_data = []
    headers = ['Method', 'Accuracy', 'F1-Score', 'AUC-ROC', 'Parameters (M)',
               'Uplink/Round (MB)', 'Downlink/Round (MB)', 'Total Comm./Round (MB)',
               'Avg. Client Time (s)', 'Round Time (s)']

    for i, (model, history, comm_costs, comp_times) in enumerate(models_results):
        final_results = history[-1]

        if hasattr(model, 'comm_details'):
            comm_details = model.comm_details
            avg_client_time = np.mean(model.client_times) if hasattr(model, 'client_times') else 0
        else:
            comm_tracker = CommCostTracker(model)
            comm_details = comm_tracker.get_detailed_costs()
            avg_client_time = np.mean(comp_times) / 4  # Estimate

        row = [
            labels[i],
            f"{final_results['accuracy']:.3f}",
            f"{final_results['f1']:.3f}",
            f"{final_results['auc_roc']:.3f}",
            f"{comm_details['parameters_m']:.1f}",
            f"{comm_details['uplink_mb']:.1f}",
            f"{comm_details['downlink_mb']:.1f}",
            f"{comm_details['total_comm_mb']:.1f}",
            f"{avg_client_time:.1f}",
            f"{np.mean(comp_times):.1f}"
        ]
        table_data.append(row)

    table = ax.table(cellText=table_data, colLabels=headers, loc='center', cellLoc='center')
    table.auto_set_font_size(False)
    table.set_fontsize(9)
    table.scale(1.2, 1.5)

    # Color the header
    for i in range(len(headers)):
        table[(0, i)].set_facecolor('#4ECDC4')
        table[(0, i)].set_text_props(weight='bold')

    # Color rows alternately
    for i in range(1, len(table_data) + 1):
        for j in range(len(headers)):
            if i % 2 == 0:
                table[(i, j)].set_facecolor('#F0F0F0')

    plt.title('Detailed Performance and Communication Analysis',
              fontsize=16, fontweight='bold', pad=20)
    plt.savefig('aptos_performance_table.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_confusion_matrix_aptos(y_true, y_pred, class_names, title="APTOS Confusion Matrix"):
    """Enhanced confusion matrix for APTOS"""
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names,
                cbar_kws={'label': 'Count'})
    plt.title(title, fontsize=16, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

# ==== Main Execution Function ====
def run_aptos_fl_experiment():
    """Run complete APTOS federated learning experiment with EfficientNetB7"""

    print("Setting up APTOS Dataset with automatic train/test split...")
    BASE_PATH = "/content/drive/MyDrive/Tanjum/Retinopathy/Extract/APTOS/gaussian_filtered_images/gaussian_filtered_images"

    TEST_RATIO = 0.2
    RANDOM_SEED = 42

    try:
        train_ds, test_ds, class_names, class_weights, num_classes = setup_aptos_dataset_with_split(
            BASE_PATH, test_ratio=TEST_RATIO, random_seed=RANDOM_SEED
        )

        test_loader = DataLoader(test_ds, batch_size=32, shuffle=False, num_workers=2)

        print("\nCreating non-IID client splits...")
        labels = np.array([y for _, y in train_ds.samples])
        client_indices = split_noniid_aptos(labels, n_clients=4, alpha=0.3)

        # Create client loaders with class balancing
        client_loaders = []
        for idx in client_indices:
            client_labels = labels[idx]
            sample_weights = class_weights[client_labels]
            sampler = WeightedRandomSampler(sample_weights, len(sample_weights), replacement=True)
            loader = DataLoader(Subset(train_ds, idx), batch_size=16, sampler=sampler, num_workers=2)
            client_loaders.append(loader)

        # Print client distribution
        print("\nClient Data Distribution:")
        for i, idx in enumerate(client_indices):
            client_labels = labels[idx]
            unique, counts = np.unique(client_labels, return_counts=True)
            class_dist = {class_names[u]: c for u, c in zip(unique, counts)}
            print(f"  Client {i+1}: {len(idx)} samples - {class_dist}")

        print("\n" + "="*80)
        print("Starting APTOS Federated Learning Experiments with EfficientNetB7")
        print("="*80)

        # Run experiments
        results = {}
        methods = ['fedavg', 'fedprox', 'moe']

        for method in methods:
            print(f"\nRunning {method.upper()} for APTOS...")

            model, history, comm, comp = run_enhanced_fl_aptos(
                mode=method,
                rounds=15,
                local_epochs=3,
                dp=False,
                lr=2e-4,  # Optimized learning rate for EfficientNetB7
                client_loaders=client_loaders,
                test_loader=test_loader,
                class_weights=class_weights,
                client_indices=client_indices,
                class_names=class_names,
                num_experts=4
            )

            results[method] = (model, history, comm, comp)

            # Print method summary
            final_results = history[-1]
            print(f"\n{method.upper()} Final Results:")
            print(f"  - Accuracy: {final_results['accuracy']:.3f}")
            print(f"  - F1-Score: {final_results['f1']:.3f}")
            print(f"  - AUC-ROC: {final_results['auc_roc']:.3f}")
            if 'expert_entropy' in final_results:
                print(f"  - Expert Entropy: {final_results['expert_entropy']:.3f}")

        # DP experiment
        print(f"\nRunning DP-FedAvg for APTOS...")
        model_dp, history_dp, comm_dp, comp_dp = run_enhanced_fl_aptos(
            mode="fedavg",
            rounds=15,
            local_epochs=3,
            dp=True,
            sigma=0.02,
            lr=2e-4,
            client_loaders=client_loaders,
            test_loader=test_loader,
            class_weights=class_weights,
            client_indices=client_indices,
            class_names=class_names
        )
        results['dp_fedavg'] = (model_dp, history_dp, comm_dp, comp_dp)

        print("\n" + "="*80)
        print("GENERATING COMPREHENSIVE ANALYSIS")
        print("="*80)

        # Visualizations
        histories = [results[method][1] for method in results.keys()]
        labels = [method.replace('_', '-').upper() for method in results.keys()]
        models_results = [results[method] for method in results.keys()]

        print("\nPlotting training metrics...")
        plot_enhanced_results_aptos(histories, labels, class_names)

        print("\nGenerating communication analysis...")
        plot_communication_analysis(models_results, labels)

        print("\nCreating detailed performance table...")
        plot_performance_table(models_results, labels, class_names)

        # Find best model
        best_method = max(results.keys(),
                         key=lambda x: max(round_data['f1'] for round_data in results[x][1]))
        best_model, best_history = results[best_method][:2]
        best_results = best_history[-1]

        print(f"\nBest Performing Method: {best_method.upper()}")
        print(f"Final Performance Metrics:")
        print(f"  - Accuracy: {best_results['accuracy']:.3f}")
        print(f"  - F1-Score: {best_results['f1']:.3f}")
        print(f"  - Recall: {best_results['recall']:.3f}")
        print(f"  - AUC-ROC: {best_results['auc_roc']:.3f}")

        # Detailed classification report
        print(f"\nDetailed Classification Report ({best_method.upper()}):")
        print(classification_report(best_results['y_true'], best_results['y_pred'],
                                  target_names=class_names, digits=3))

        # Confusion Matrix
        print("\nGenerating confusion matrix...")
        plot_confusion_matrix_aptos(
            best_results['y_true'],
            best_results['y_pred'],
            class_names,
            f"APTOS Confusion Matrix - {best_method.upper()}"
        )

        # Grad-CAM Visualizations
        print("\nGenerating Grad-CAM visualizations...")
        generate_gradcam_visualization(best_model, test_loader, class_names, num_samples=5)

        # ROC Curves
        print("\nMulti-class ROC Analysis...")
        plt.figure(figsize=(12, 8))
        y_true_bin = label_binarize(best_results['y_true'], classes=range(num_classes))
        colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#F39C12']

        for i, (class_name, color) in enumerate(zip(class_names, colors)):
            if len(np.unique(y_true_bin[:, i])) > 1:
                fpr, tpr, _ = roc_curve(y_true_bin[:, i], best_results['y_prob'][:, i])
                auc_score = auc(fpr, tpr)
                plt.plot(fpr, tpr, color=color, linewidth=2,
                        label=f'{class_name} (AUC = {auc_score:.3f})')

        plt.plot([0, 1], [0, 1], 'k--', alpha=0.5)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Multi-class ROC Curves - APTOS Dataset (EfficientNetB7)')
        plt.legend(loc="lower right")
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

        # Final Summary
        print("\n" + "="*80)
        print("APTOS FEDERATED LEARNING EXPERIMENT COMPLETED!")
        print("="*80)

        print(f"\nKey Findings:")
        print(f"  - Best Method: {best_method.upper()}")
        print(f"  - Model: EfficientNetB7 with federated learning")
        print(f"  - Dataset: APTOS (5-class diabetic retinopathy)")
        print(f"  - Clients: 4 with non-IID data distribution")
        print(f"  - Best F1-Score: {best_results['f1']:.3f}")
        print(f"  - Best Accuracy: {best_results['accuracy']:.3f}")
        print(f"  - Multi-class AUC: {best_results['auc_roc']:.3f}")

        if hasattr(best_model, 'comm_details'):
            comm_details = best_model.comm_details
            print(f"\nCommunication Metrics:")
            print(f"  - Parameters (M): {comm_details['parameters_m']:.1f}")
            print(f"  - Uplink/Round (MB): {comm_details['uplink_mb']:.1f}")
            print(f"  - Downlink/Round (MB): {comm_details['downlink_mb']:.1f}")
            print(f"  - Total Comm./Round (MB): {comm_details['total_comm_mb']:.1f}")
            if hasattr(best_model, 'client_times'):
                print(f"  - Avg. Client Time (s): {np.mean(best_model.client_times):.1f}")

        print(f"\nTechnical Achievements:")
        print(f"  - Successfully implemented EfficientNetB7 for FL-MOE")
        print(f"  - Achieved improved performance over DenseNet169")
        print(f"  - Comprehensive communication cost analysis")
        print(f"  - Working Grad-CAM visualizations for interpretability")
        print(f"  - Detailed timing and efficiency metrics")
        print(f"  - Enhanced class imbalance handling")

        return results, best_model, class_names

    except Exception as e:
        print(f"Error in experiment: {str(e)}")
        print("Troubleshooting tips:")
        print("1. Check dataset path is correct")
        print(f"   Current path: {BASE_PATH}")
        print("2. Ensure sufficient GPU memory for EfficientNetB7")
        print("3. Verify all dependencies are installed")
        import traceback
        traceback.print_exc()
        return None, None, None

# Run the experiment
if __name__ == "__main__":
    results, best_model, class_names = run_aptos_fl_experiment()
```
